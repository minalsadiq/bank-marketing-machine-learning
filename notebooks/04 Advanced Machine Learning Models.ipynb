Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

import warnings
warnings.filterwarnings("ignore")
```

Load Final Machine Learning Dataset

```python
df = pd.read_csv("../src/data/processed/bank_ml_dataset.csv")

print("Dataset Loaded Successfully.\n")
print("Dataset Shape :", df.shape)

df.head()
```

Split Features and Target

```python
X = df.drop("y", axis=1)
y = df["y"]

print("Features Shape :", X.shape)
print("Target Shape   :", y.shape)
```

Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)

print("Training Target :", y_train.shape)
print("Testing Target  :", y_test.shape)
```

Verify Dataset

```python
print("Advanced Machine Learning Dataset Ready")

print("Training Samples :", len(X_train))
print("Testing Samples  :", len(X_test))

print("\nTarget Distribution")
print(y.value_counts())
```

XGBoost Classifier

```python
xgb_model = XGBClassifier(
    random_state=42,
    eval_metric="logloss"
)

xgb_model.fit(X_train, y_train)

print("XGBoost Model Trained Successfully.")
```

XGBoost Evaluation

```python
xgb_pred = xgb_model.predict(X_test)

accuracy_xgb = accuracy_score(y_test, xgb_pred) * 100
precision_xgb = precision_score(y_test, xgb_pred) * 100
recall_xgb = recall_score(y_test, xgb_pred) * 100
f1_xgb = f1_score(y_test, xgb_pred) * 100
roc_auc_xgb = roc_auc_score(y_test, xgb_pred) * 100

print(f"Accuracy : {accuracy_xgb:.2f}%")
print(f"Precision: {precision_xgb:.2f}%")
print(f"Recall   : {recall_xgb:.2f}%")
print(f"F1 Score : {f1_xgb:.2f}%")
print(f"ROC AUC  : {roc_auc_xgb:.2f}%")

print("\nClassification Report")
print(classification_report(y_test, xgb_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, xgb_pred))
```

LightGBM Classifier

```python
lgb_model = LGBMClassifier(
    random_state=42
)

lgb_model.fit(X_train, y_train)

print("LightGBM Model Trained Successfully.")
```

LightGBM Evaluation

```python
lgb_pred = lgb_model.predict(X_test)

accuracy_lgb = accuracy_score(y_test, lgb_pred) * 100
precision_lgb = precision_score(y_test, lgb_pred) * 100
recall_lgb = recall_score(y_test, lgb_pred) * 100
f1_lgb = f1_score(y_test, lgb_pred) * 100
roc_auc_lgb = roc_auc_score(y_test, lgb_pred) * 100

print(f"Accuracy : {accuracy_lgb:.2f}%")
print(f"Precision: {precision_lgb:.2f}%")
print(f"Recall   : {recall_lgb:.2f}%")
print(f"F1 Score : {f1_lgb:.2f}%")
print(f"ROC AUC  : {roc_auc_lgb:.2f}%")

print("\nClassification Report")
print(classification_report(y_test, lgb_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, lgb_pred))
```

CatBoost Classifier

```python
cat_model = CatBoostClassifier(
    random_state=42,
    verbose=0
)

cat_model.fit(X_train, y_train)

print("CatBoost Model Trained Successfully.")
```

CatBoost Evaluation

```python
cat_pred = cat_model.predict(X_test)

accuracy_cat = accuracy_score(y_test, cat_pred) * 100
precision_cat = precision_score(y_test, cat_pred) * 100
recall_cat = recall_score(y_test, cat_pred) * 100
f1_cat = f1_score(y_test, cat_pred) * 100
roc_auc_cat = roc_auc_score(y_test, cat_pred) * 100

print(f"Accuracy : {accuracy_cat:.2f}%")
print(f"Precision: {precision_cat:.2f}%")
print(f"Recall   : {recall_cat:.2f}%")
print(f"F1 Score : {f1_cat:.2f}%")
print(f"ROC AUC  : {roc_auc_cat:.2f}%")

print("\nClassification Report")
print(classification_report(y_test, cat_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, cat_pred))
```

Advanced Models Comparison

```python
comparison = pd.DataFrame({

    "Model":[
        "XGBoost",
        "LightGBM",
        "CatBoost"
    ],

    "Accuracy (%)":[
        accuracy_xgb,
        accuracy_lgb,
        accuracy_cat
    ],

    "Precision (%)":[
        precision_xgb,
        precision_lgb,
        precision_cat
    ],

    "Recall (%)":[
        recall_xgb,
        recall_lgb,
        recall_cat
    ],

    "F1 Score (%)":[
        f1_xgb,
        f1_lgb,
        f1_cat
    ],

    "ROC AUC (%)":[
        roc_auc_xgb,
        roc_auc_lgb,
        roc_auc_cat
    ]

})

comparison
```

Save Comparison Results

```python
comparison.to_csv(
    "../reports/model_comparison_advanced.csv",
    index=False
)

print("Advanced Model Comparison Saved Successfully.")
```

Best Advanced Model

```python
best_model = comparison.loc[
    comparison["Accuracy (%)"].idxmax()
]

print("BEST ADVANCED MODEL")

print(best_model)
```
