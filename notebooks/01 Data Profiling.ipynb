```python

# Basic Dataset Inspection
# =====================================================

# Dataset shape
print("Dataset Shape")
print(df.shape)

# -----------------------------------------------------

# First five rows
print("\nFirst Five Rows")
display(df.head())

# -----------------------------------------------------

# Last five rows
print("\nLast Five Rows")
display(df.tail())

# -----------------------------------------------------

# Column names
print("\nColumn Names")
print(df.columns.tolist())

# -----------------------------------------------------

# Data types
print("\nData Types")
display(df.dtypes)

# -----------------------------------------------------

# Dataset information
print("\nDataset Information")
df.info()

# -----------------------------------------------------

# Missing values
print("\nMissing Values")
display(df.isnull().sum())

# -----------------------------------------------------

# Duplicate rows
print("\nDuplicate Rows")
print(df.duplicated().sum())
```

```python
# Target Variable Analysis
# =====================================================

import matplotlib.pyplot as plt
import seaborn as sns

# Count of each class
target_counts = df["y"].value_counts()

print("Target Variable Counts")
print(target_counts)

# Percentage
target_percentage = round(df["y"].value_counts(normalize=True) * 100, 2)

print("Target Variable Percentage")
print(target_percentage)

# -----------------------------------------------------
# Visualization
# -----------------------------------------------------

sns.countplot(
    data=df,
    x="y",
    hue="y",
    palette="Set2",
    legend=False
)

plt.title("Target Variable Distribution")
plt.xlabel("Subscribed to Term Deposit")
plt.ylabel("Count")

plt.show()

# -----------------------------------------------------
# Pie Chart
# -----------------------------------------------------

plt.figure(figsize=(6,6))

plt.pie(
    target_counts,
    labels=target_counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Target Variable Percentage")

plt.show()
```

```python
# =====================================================
# Numerical Features Distribution
# =====================================================

import matplotlib.pyplot as plt
import seaborn as sns

# Numerical columns
numerical_columns = [
    "age",
    "duration",
    "campaign",
    "pdays",
    "previous",
    "emp.var.rate",
    "cons.price.idx",
    "cons.conf.idx",
    "euribor3m",
    "nr.employed"
]

# Create figure
plt.figure(figsize=(20, 15))

# Plot histogram for each numerical feature
for i, column in enumerate(numerical_columns):

    plt.subplot(4, 3, i + 1)

    sns.histplot(
        data=df,
        x=column,
        bins=30,
        kde=True,
        color="steelblue"
    )

    plt.title(f"Distribution of {column}")

plt.tight_layout()
plt.show()
```

```python
# =====================================================
# Categorical Features Distribution
# =====================================================

import matplotlib.pyplot as plt
import seaborn as sns

# Select categorical columns
categorical_columns = df.select_dtypes(include="object").columns

# Exclude target variable
categorical_columns = categorical_columns.drop("y")

# Plot each categorical feature
for column in categorical_columns:

    plt.figure(figsize=(8,4))

    sns.countplot(
        data=df,
        x=column,
        hue=column,
        palette="Set2",
        legend=False,
        order=df[column].value_counts().index
    )

    plt.title(f"Distribution of {column}")
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
```
